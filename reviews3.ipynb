```python
new_price=[]
for items in df['Price'].str.findall(r'\d+\.*\d*'):
    new_price.append(float(items[0]))
df['Price']=new_price
```

```python
new_items=[]
for items in list(df['Installs'].str.findall(r'\d+,?\d*,?\d*,?\d*,?\d*,?\d*,?\d*')):
    new_items.append((items[0]))
new_new_items=[]
for items in new_items:
    if ',' in items:
        new_new_items.append(float(items.replace(",", "")))
    if ',' not in items:
        new_new_items.append(float(items))
df['Installs']=new_new_items
```

```python
# Type encoding
df['Type'] = df['Type'].map({
    'Free' : 0, 
    'Paid': 1
})
```
